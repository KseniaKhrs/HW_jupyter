# f(x) = sin(x)^2 - cos(x)^2

```python
from sympy.plotting import plot
from sympy import *
x = Symbol('x')
function = sin(x)**2 - cos(x)**2
```

## 1. Определить корни

```python
answer = solve(function,x)
print(answer)
for i in answer:
    print(i.evalf())
```

```python
solveset(function,x)
```

## 2. Найти интервалы, на которых функция возрастает

```python
deriv = diff(function,x)
solve(deriv > 0, x)
```

## 3. Найти интервалы, на которых функция убывает

```python
solve(deriv < 0, x)
```

## 4. Построить график

```python
chart = plot(function, show=False)
chart.show()
```
